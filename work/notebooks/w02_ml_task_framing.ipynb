## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking / Scoring.

I chose ranking because my aim is to identify which pages should be reviewed first. The model would give pages a score based on their performance, so the pages with the biggest opportunity can be checked first. This is better than treating every page the same because some pages need more attention than others.
