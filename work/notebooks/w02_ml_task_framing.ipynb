## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking / Scoring.

I chose ranking because my aim is to identify which pages should be reviewed first. The model would give pages a score based on their performance, so the pages with the biggest opportunity can be checked first. This is better than treating every page the same because some pages need more attention than others.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

I want to score pages based on how much attention they may need. Since the starter dataset does not contain future outcomes, I will use the current CTR together with search position as a proxy. Pages with lower-than-expected CTR for their position can be given a higher priority for review.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

I would measure success by how accurately the model identifies pages with low CTR for their search position. A good result would be that most of the pages flagged by the model are truly pages that need review, allowing the SEO team to spend time on the right pages instead of checking every page manually.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule, such as "review every page with CTR below 0.2%", is too simple because CTR depends on many factors, including search position, impressions, content type, and search intent. The relationship between these factors is not always the same. An ML model can learn these patterns from the data and rank pages that are most likely to need review better than a single fixed rule.
